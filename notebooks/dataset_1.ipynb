# Exploration du jeu de données : *ccfraud_dataset1.csv*

## Explication du dataset

Ce jeu de données regroupe les transactions bancaires effectuées en septembre 2013 par des citoyens européens détenteurs d'une carte de crédit. Toutes ces transactions ont ainsi eu lieu en l'espace de deux jours.

Par soucis de confidentialité, les données concernant les clients des banques n'ont pas pu être révélés dans le jeu de données. Ainsi, elles ont été remplacés par les composantes principales `V1`,`V2`, ...,`V28` par transformation par ACP.

Les seules colonnes non transformées sont :

* `Time` : Le temps passé (*en secondes*) depuis la première transaction enregistrée dans le jeu de données.
* `Amount` : Le montant de la transaction (*en €*)
* `Class` : Un booléen qui indique si la transaction est une fraude ou non. **1** si c'est une fraude, **0** si elle est légitime.
